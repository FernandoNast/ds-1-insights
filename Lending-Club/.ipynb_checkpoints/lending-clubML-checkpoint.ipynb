```python
df1 = data_raw.copy()
dt1 = pd.to_datetime(df1['issue_d'])
df1['year'] = dt1.dt.year
```

```python
df1.head()
```

```python
df1['complete_date'] = pd.to_datetime(df1['issue_d'])
df1['complete_date'].head()
```

```python
# Agrupando os estados em regioes, fazendo assim diminuir a quantidade de valores
df2 = df1.copy()

oeste = ['CA', 'OR', 'UT','WA', 'CO', 'NV', 'AK', 'MT', 'HI', 'WY', 'ID']
sudoeste = ['AZ', 'TX', 'NM', 'OK']
sudeste = ['GA', 'NC', 'VA', 'FL', 'KY', 'SC', 'LA', 'AL', 'WV', 'DC', 'AR', 'DE', 'MS', 'TN' ]
centro_oeste = ['IL', 'MO', 'MN', 'OH', 'WI', 'KS', 'MI', 'SD', 'IA', 'NE', 'IN', 'ND']
nordeste = ['CT', 'NY', 'PA', 'NJ', 'RI','MA', 'MD', 'VT', 'NH', 'ME']

def finding_regions(state):
    if state in oeste:
        return 'Oeste'
    elif state in sudoeste:
        return 'Sudoeste'
    elif state in sudeste:
        return 'Sudeste'
    elif state in centro_oeste:
        return 'Centro_oeste'
    elif state in nordeste:
        return 'Nordeste'
    
df2['regiao'] = df2['addr_state'].apply(finding_regions)
```

```python
df2.head()
```

## Verificando os NaNs

```python
# len(data_raw) = 248196

colunas = data_raw.columns
df3 = df2.copy()

count = 0
for col in colunas:
    if (df3[str(col)].isnull().sum()>=21000): # Selecionei 21000 pq acima disso tem-se 99% nulos, abaixo disso o que mais tem nulos, chega a 8%
        print(str(col))
        print(df3[str(col)].isnull().sum())
        print(df3[str(col)].isnull().sum()/len(df3))
        print('\n')
        count=count+1
        
print('Quantidade de colunas com mais de 21k NaN -',count)
```

```python
df3 = df3.dropna(axis=1,thresh=(0.0846*len(data_raw)))
```

```python
df3.head()
```

```python
# Agrupar Default, Late em 1 e Current em 0. Ou seja, 1 Ã© para os que atrasaram e 0 Ã© o que estÃ¡ em dia

df4 = df3.copy()

def troca_valores(row):
    if((row['loan_status']=='Default')|(row['loan_status']=='Late (31-120 days)')|(row['loan_status']=='Late (16-30 days)')):
        return 1
    else:
        return 0
    
df4['loan_status'] = df3.apply(df3,axis=1)
```

```python
for col in df4.columns.values:
    if (df4[str(col)].isnull().sum()>0): 
        print(str(col))
        print('Qtd Absoluta:',df4[str(col)].isnull().sum())
        print('Qtd %:', df4[str(col)].isnull().sum()/len(df4))
        print('Media:',df4[str(col)].mean())
        print('Mediana:',df4[str(col)].median())
        print('\n')
```
